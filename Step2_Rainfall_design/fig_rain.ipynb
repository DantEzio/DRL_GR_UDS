# 做降雨相关图

```python
import numpy as np
import Rainfall_data as RD
import matplotlib.pyplot as plt
```

```python
rainfalle = np.load('./RSH/east.npy').tolist()
rainfallw = np.load('./RSH/west.npy').tolist()
re,rw={},{}
for i in range(len(rainfalle)):
    teme,temw=[],[]
    for j in range(len(rainfalle[i])):
        teme.append(float(rainfalle[i][j][1]))
        temw.append(float(rainfallw[i][j][1]))
    re['rain'+str(i)]=teme
    rw['rain'+str(i)]=temw
```

# Fig

```python
font1={
    'family':'serif',
    'size':13,
}

font2={
    'family':'serif',
    'size':20,
}

fig,ax=plt.subplots(6,2,figsize=(10,8))
plt.subplots_adjust(hspace=0.4)
x=np.arange(0,120)
for i in range(6):
    ax[i,0].bar(x,re['rain'+str(i)])
    ax[i,1].bar(x,rw['rain'+str(i)])
    ax[i,0].set_title('RSH'+str(i+1)+': East rainfall',font=font1)
    ax[i,1].set_title('RSH'+str(i+1)+': West rainfall',font=font1)

    if i == 2:
        ax[i,0].set_ylabel('Rain intensity (mm/hour)',font=font2)
        ax[i,1].set_ylabel('Rain intensity (mm/hour)',font=font2)

    if i == 5:
        ax[i,0].set_xlabel('Time (minutes)',font=font1)
        ax[i,1].set_xlabel('Time (minutes)',font=font1)
    else:
        ax[i,0].set_xticks([])
        ax[i,1].set_xticks([])

plt.savefig('rainfall_fig.tif',dpi=300,bbox_inches='tight')
```
