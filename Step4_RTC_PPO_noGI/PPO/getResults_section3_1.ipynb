```python
# 读取顺序
results={}
#For RSH
tem={
    'PPO': [np.load('./Results/RSH/'+str(i)+'.npy',allow_pickle=True).tolist() for i in range(6)],
            }
results['RSH']=tem

#For RR
tem={
    'PPO': [np.load('./Results/RR/'+str(i)+'.npy',allow_pickle=True).tolist() for i in range(4)],
            }
results['RR']=tem
```

```python
re=[]
for item in results['RSH'].keys():#['Do'+'\n'+'nothing','GR','HC','DDQN','PPO']
    tem=[]
    for rain in range(6):
        rpt=results['RSH'][item][rain]
        for jt in range(2):
            rpt['RES'][jt]=1.0
        tem = [rpt['F'][-1]+rpt['C'][-1],np.mean(rpt['RES'])]
        re.append(tem)
print(pd.DataFrame(re))
```
